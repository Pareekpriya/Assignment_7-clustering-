# Hierarchical Clustering

```python
from sklearn.preprocessing import MinMaxScaler
trans = MinMaxScaler()
data1 = pd.DataFrame(trans.fit_transform(data.iloc[:,1:]))
data1  
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x) 
```

```python
df_norm = norm_func(data.iloc[:,1:])
df_norm
```

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch # for creating dendrogram 
#p = np.array(df_norm) # converting into numpy array format 
z = linkage(df_norm, method="average",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
    z,
    #leaf_rotation=6.,  # rotates the x axis labels
    #leaf_font_size=15.,  # font size for the x axis labels
)
plt.show()  
```

```python
from sklearn.cluster import AgglomerativeClustering 
import warnings 
warnings.filterwarnings('ignore')
h_complete = AgglomerativeClustering(n_clusters=4, linkage='average',affinity = "euclidean").fit(df_norm) 

cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
data['clust']=cluster_labels # creating a  new column and assigning it to new column 
data 
```

# DB Scan Clustering

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
# Import .csv file and convert it to a DataFrame object
df = pd.read_csv("crime_data.csv");
print(df.head())
df 
```

```python
print(data.info())
```

```python
df1 = df.drop(['Unnamed: 0'],axis=1)
df1
```

```python
array=df1.values
array
```

```python
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array) 
X   
```

```python
dbscan = DBSCAN(eps=1, min_samples=5)
dbscan.fit(X)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_ 
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])  
```

```python
cl
pd.set_option("display.max_rows", None)  
```

```python
cl   
```

```python
df1 = pd.concat([df,cl],axis=1)  
df1     
```

```python
import matplotlib.pyplot as plt
plt.style.use('classic')   
```

```python
df1.plot(x="UrbanPop",y ="Rape",c=dbscan.labels_ ,kind="scatter",s=50 ,cmap=plt.cm.copper_r) 
plt.title('Clusters using DBScan') 
```

```python
dl = dbscan.labels_ 
dl
```

```python
import sklearn
sklearn.metrics.silhouette_score(X,dl)
```

```python
from sklearn.cluster import KMeans
clf = KMeans(n_clusters=3)
y_kmeans = clf.fit_predict(X) 
```

```python
y_kmeans
```

```python
cl1=pd.DataFrame(y_kmeans,columns=['Kcluster']) 
cl1 
```

```python
df2 = pd.concat([df1,cl1],axis=1) 
df2 
```

```python
df2.plot(x="UrbanPop",y ="Rape",c=y_kmeans ,kind="scatter",s=50 ,cmap=plt.cm.copper_r) 
plt.title('Clusters using KMeans')  
```

```python
sklearn.metrics.silhouette_score(X, y_kmeans)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
neigh = NearestNeighbors(n_neighbors=8)
nbrs = neigh.fit(X)
distances, indices = nbrs.kneighbors(X)
```
